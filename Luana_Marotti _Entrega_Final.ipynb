1. Descargar, leer y unir en una sola estructura de datos. ¿Qué estructura elegiste y por qué?

````python
#Abro archivos
path_to_file_arg="2021.2 - sysarmy - Encuesta de remuneración salarial Argentina - Argentina.csv"
path_to_file_latam="2021.2 - sysarmy - Encuesta de remuneración salarial Latam - Latam.csv"

data_arg=[]
data_latam=[]

import csv

with open(path_to_file_arg, 'r',encoding='utf8') as path_to_file_arg:
    reader=csv.reader(path_to_file_arg)
    for row in reader:
        data_arg.append(row[:])

with open(path_to_file_latam, 'r',encoding='utf8') as path_to_file_latam:
    reader=csv.reader(path_to_file_latam)
    for row in reader:
        data_latam.append(row[:])

#Elimino las filas innecesarias 
del data_arg[:9]
del data_latam[:11]

#Elimino columnas extras en el archivo de latam
for column in data_latam:
    i=18
    while i<=18 and i>0:
        if len(column[i])==0:
            column.pop(i)
        i-=1
            
#Agrego columna país en el archivo de argentina
data_arg[0].insert(0,'En que pais trabajas')
for i in range(1,len(data_arg),1):
    data_arg[i].insert(0,'Argentina')
    
#Junto los conjuntos de datos
data=data_arg+data_latam

#Creo un diccionario de listas con las preguntas como keys y una lista con las respuestas como elemento
dict_data={}
for i in range (0,len(data[0]),1):
    dict_data.update({data[0][i]:[]})
    for j in range (1,len(data),1):
        data[j][i]=data[j][i].lower()
        data[j][i]=data[j][i].replace('á','a').replace('é','e').replace('í','i').replace('ó','o').replace('ú','u').replace('ñ','n')
        data[j][i]=data[j][i].replace('lic ','licenciatura ').replace('lic. ','licenciatura ')
        data[j][i]=data[j][i].replace('ed ','educacion').replace('ed. ','educacion')
        data[j][i]=data[j][i].replace('tec ','tecnicatura').replace('tec. ','tecnicatura')
        data[j][i]=data[j][i].replace('cs ','ciencias').replace('cs. ','ciencias')
        data[j][i]=data[j][i].replace('cr ','contador').replace('cr. ','contador')
        data[j][i]=data[j][i].replace('arq','arquitecto').replace('arq. ','arquitecto')
        data[j][i]=data[j][i].replace('mg ','magister').replace('mg. ','magister')
        data[j][i]=data[j][i].replace(',','.')
        dict_data[data[0][i]].append(data[j][i])

#Cierro archivos
path_to_file_arg.close()
path_to_file_latam.close()

print('Elijo un diccionario de listas porque considero que me permite acceder a las respuetas de los participantes con facilidad')
````

2. Printear porcentaje de participación por región, ordenado de mayor a menor. Ej: 

```
- Ciudad Autónoma de Buenos Aires - 59.4%
- Formosa - 0.04%
```

````python
#Filtro las personas que no respondieron esta pregunta
auxiliar_region=dict_data[data[0][1]] #Creo una lista auxiliar en donde puedo eliminar los espacios vacíos
while '' in auxiliar_region:
    auxiliar_region.remove('')
    
#Obtengo una lista con todas las regiones
set_region=set(auxiliar_region) 
list_region=list(set_region)

#Creo una función para obtener un diccionario con las respuestas, cantidad de veces que aparece y %
def dict_answers_criteria(list_a,list_b,list_c):
    dict_criteria={}
    for i in range(0,len(list_a),1): 
        dict_criteria.update({i:[list_a[i],0,0]})
    for i in range(0,len(list_b),1):
        for j in range(0,len(list_a),1):
                   if list_b[i]==dict_criteria[j][0]:
                       dict_criteria[j][1]+=1
    for i in range(0,len(list_a),1):
        dict_criteria[i][2]=round(dict_criteria[i][1]*100/(len(list_c)),2)
    return dict_criteria

#Obtengo un diccionario con las regiones, cantidad de veces que aparece y %
dict_region=dict_answers_criteria(list_region,dict_data[data[0][1]],auxiliar_region)

#Creo una función para ordenar según % de mayor a menor
def order_percentage_criteria(list_a,dict_b):
    count=len(list_a)
    used_list=[]
    while count>0:
        max_region_name=[]
        max_region_percentage=0
        for i in range(0,len(list_a),1):
            if dict_b[i][2]>max_region_percentage and dict_b[i][0] not in used_list:
                max_region_percentage=dict_b[i][2]
                max_region_name=dict_b[i][0]
        print(f'{max_region_name.capitalize()} - {max_region_percentage}%')
        used_list.append(max_region_name)
        count=count-1

#Ordeno las regiones segun sus % de mayor a menor
order_percentage_region=order_percentage_criteria(list_region,dict_region)
order_percentage_region
````

````python
#Creo una función para obtener la mediana del salario neto según un parámetro
def median_salary_criteria(list_a,list_b,dict_salary_criteria):
    for i in range(0,len(list_a),1): 
        dict_salary_criteria.update({i:[list_a[i],[],0]})
    for i in range(0,len(list_b),1):
        for j in range(0,len(list_a),1):
               if list_b[i]==dict_salary_criteria[j][0]:
                   if type(dict_data[data[0][4]][i])==float:
                       dict_salary_criteria[j][1].append(dict_data[data[0][4]][i])
    import statistics
    for i in range(0,len(list_a),1):
        if dict_salary_criteria[i][1]!=[]:
            dict_salary_criteria[i][2]=round(statistics.median(dict_salary_criteria[i][1]))
    
    #Ordeno según mediana de mayor a menor
    count=len(list_a)
    used_list=[]
    while count>0:
        max_name=[]
        max_median=0
        for i in range(0,len(list_a),1):
            if dict_salary_criteria[i][2]>max_median and dict_salary_criteria[i][0] not in used_list:
                max_median=dict_salary_criteria[i][2]
                max_name=dict_salary_criteria[i][0]
        if max_median>200:
            print(f'{max_name.capitalize()} - {max_median} USD')
        used_list.append(max_name)
        count=count-1
````
